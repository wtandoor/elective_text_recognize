```python
df = pd.read_csv('/content/drive/MyDrive/elective_text_recognize/ЛР2/my_submission.csv', sep=',')
df
```

```python
dictionary_lemma = pd.read_csv('/content/drive/MyDrive/elective_text_recognize/ЛР3/Dictionary_lemmatization_Rus (1).csv', sep= ',')
dictionary_lemma
```

```python
vector_list_all = []
for i in range(535):
  string = df['Transcription'].values[i].split(' ')
  vector_list = []
  for k in string:
    for l in range(19637):
      if dictionary_lemma['Word'].values[l] == k:
        vector_list.append(dictionary_lemma.loc[l, 'Weight'])
        break
      if l == 19636:
        vector_list.append(0)
  vector_list_all.append(vector_list)
```

```python
vector_list_all[0]
```

Далее из полученных весов находим следующие параметры и создаем из них вектор.

1. Кол-во положительных слов.
2. Кол-во отрицаильных слов.
3. Среднее значение сентимента всех слов.
4. Максмально положительное значение вектора.
5. Максимально отрицательное значение вектора.

```python
static_vector = []

for i in vector_list_all:
  static_vector_each = []
  positive_weights = 0
  negative_weights = 0
  avg_sentiment = 0
  count_of_words = len(i)
  max_positive = 0
  max_negative = 0
  for k in i:
    if k > 0:
      positive_weights += 1
    elif k < 0:
      negative_weights += 1
    avg_sentiment += k
    if k > max_positive:
      max_positive = k
    if k < max_negative:
      max_negative = k
  static_vector_each.append(positive_weights)
  static_vector_each.append(negative_weights)
  static_vector_each.append(avg_sentiment/count_of_words)
  static_vector_each.append(max_positive)
  static_vector_each.append(max_negative)
  static_vector.append(static_vector_each)


for i in static_vector:
  print(i)
```

```python
tone_dictionary = pd.DataFrame(static_vector, columns = ['positive', 'negative', 'avg_sentiment', 'max_positive', 'max_negative'])
tone_dictionary
```

Мы векторизовали все текста при помощи тонального словаря

Bow


```python
vectorizer = CountVectorizer()
```

```python
corpus = []
for i in range(535):
  string = df['Transcription'].values[i]
  corpus.append(string)

for i in corpus:
  print(i)
```

```python
X = vectorizer.fit_transform(corpus).toarray()
bow = pd.DataFrame(X, columns = vectorizer.get_feature_names_out())
bow.tail(6)
```

TFIDF

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus).toarray()
print(X)
```

```python
tfidf = pd.DataFrame(X, columns = vectorizer.get_feature_names_out())
tfidf
```
