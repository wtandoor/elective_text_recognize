```python
import pandas as pd
import numpy as np
import nltk
import pymorphy2
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import word_tokenize
from string import punctuation
nltk.download('punkt')
nltk.download('stopwords')
```

```python
input_string = 'В прошлом месяце преступник осуществил захват морского судна в Суэцком канале. Новость об этом преступлении шокировала многих людей, в том числе вызвала гнев у международного сообщества.'
df = pd.read_csv('/content/drive/MyDrive/elective_text_recognize/ЛР1/my_submission.csv', sep= ',')
```

```python
punctuations = list(punctuation)
stopwords = nltk.corpus.stopwords.words('russian')
morph = pymorphy2.MorphAnalyzer()
```

```python
def concat_words(argv1, argv2):
  return argv1 + " " + argv2

def preproccesing_text(input_string, punctuations=punctuations, stopwords=stopwords, morph=morph):
  tokens = word_tokenize(input_string)
  tokens_without_punc = []
  #избавляемся от пунктуации
  for i in tokens:
    if i not in punctuations:
      tokens_without_punc.append(i.lower())

  tokens_without_stopwords = []
  #избавляемся от стоп слов
  for i in tokens_without_punc:
    if i not in stopwords:
      tokens_without_stopwords.append(i)


  lemmatization_form = []
  #лемматизируем слова
  for i in tokens_without_stopwords:
      lemmatization_form.append(morph.parse(i)[0].normal_form)

  output_string = ""
  #конкатинируем слова в одну строку
  for i in lemmatization_form:
    output_string = concat_words(output_string, i)

  #обрезаем пробел в начале  
  return output_string.strip(' ')
```

```python
#изменяем поле транскипции во всем файле, используя функцию препроцессинга
for i in range(535):
  df.loc[i, 'Transcription'] = preproccesing_text(df['Transcription'].values[i])
```
