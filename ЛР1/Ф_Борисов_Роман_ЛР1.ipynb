```python
df.drop(['Name_audio'],inplace=True, axis=1)
```

```python
df.rename(columns={'label':'emotion'}, inplace=True)
df
```

```python
df['Sentiment'] = ' '
```

```python
sentiment_digits = [0, #negative
                  1,  #neutral
                  2]  #positive
```

```python
np.unique(df['emotion'])
```

```python
emotion_digits = [0, #гнев
                  1, #грусть
                  2, #отвращение
                  3, #страх
                  4, #нейтраль
                  5, #радость
                  6] #удивление
```

```python
k, l = df.shape #узнаем размерность набора данных
k
```

```python
#создаем списки с эмоциями
emotion_list = np.unique(df['emotion'])



negative_emotion = ['гнев',
                  'грусть',
                  'отвращение',
                  'страх']


neutral_emotion = ['нейтраль']



positive_emotion = ['радость',
                    'удивление']
```

Сразу заполняем три новые колонки, проходя по датафрейму с помощью цикла.

```python
df['sentiment_digits'] = np.zeros(k)
df['emotion_digits'] = np.zeros(k)
for i in range(k):
  if df['emotion'].values[i] in negative_emotion:
    df.loc[i, 'Sentiment'] = 'негативный'
    df.loc[i, 'sentiment_digits'] = sentiment_digits[0]
    if df['emotion'].values[i] == negative_emotion[0]:
      df.loc[i, 'emotion_digits'] = emotion_digits[0]
    elif df['emotion'].values[i] == negative_emotion[1]:
      df.loc[i, 'emotion_digits'] = emotion_digits[1]
    elif df['emotion'].values[i] == negative_emotion[2]:
      df.loc[i, 'emotion_digits'] = emotion_digits[2]
    else:
      df.loc[i, 'emotion_digits'] = emotion_digits[3]
  elif df['emotion'].values[i] in neutral_emotion:
    df.loc[i,'Sentiment'] = 'нейтральный'
    df.loc[i,'sentiment_digits'] = sentiment_digits[1]
    df.loc[i, 'emotion_digits'] = emotion_digits[4]
  else:
    df.loc[i,'Sentiment'] = 'позитивный'
    df.loc[i,'sentiment_digits'] = sentiment_digits[2]
    if df['emotion'].values[i] == positive_emotion[0]:
      df.loc[i, 'emotion_digits'] = emotion_digits[5]
    elif df['emotion'].values[i] == positive_emotion[1]:
      df.loc[i, 'emotion_digits'] = emotion_digits[6]
```
